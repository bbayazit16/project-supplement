```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def factorial(num):
  return 1 if num == 0 else num * factorial(num - 1)
```

```python
def bbp(num):
  a = factorial(364)
  b = factorial(365 - num)
  c = 365 ** (num - 1)
  return 1 - (a / (b * c))
```

```python
def gc(num):
  x = np.arange(1, num+1)
  y = [bbp(int(i)) for i in x]
  sns.set()
  return plt.plot(x, y)
```

```python
gc(50)
```

```python
def e(i):
  if i < 153:
    return (np.tanh(i) + 1) / 2
  return (np.tanh(i) + 1) / 2
```

```python
def similarity(num):
  x = bbp(num)
  y = e(num)
  return x - y
```

```python
# plt.ylim(-0.001, 0.001)
# plt.yticks([-0.001, -0.000, 0.001])
plt.plot([similarity(i)*-1 for i in range(2, 365)])
```
