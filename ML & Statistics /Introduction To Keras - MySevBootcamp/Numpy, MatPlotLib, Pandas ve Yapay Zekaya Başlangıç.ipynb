## Aktivasyonlar:

Yapay zeka'da aktivasyonları Koç Üniversitesi'nin derslerinde görmüştük. Özetlemek gerekirse:

> So what does an artificial neuron do? Simply put, it calculates a “weighted sum” of its input, adds a bias and then decides whether it should be “fired” or not. We decided to add “activation functions” for this purpose. To check the Y value produced by a neuron and decide whether outside connections should consider this neuron as “fired” or not. Or rather let’s say — “activated” or not.

Aktivasyon fonksiyonları, bir nöronun aktive olup olmayacağına karar vermede yardım eden fonksiyonlar. Birçok aktivasyon fonksiyou var, hepsinin de iyi ve kötü yanları var. Aktivasyon fonksiyonlarını özetlemek için iki kategoriye ayırabiliriz:

### Hidden Layer Aktivasyon Fonksiyonları:

En çok kullanılan iki hidden layer aktivasyonları aşağıdaki gibidir:

> Rectified Linear Unit (RELU): Relu zamanla tercih edilen bir fonksiyon haline gelmiştir. Eğer input 0'dan küçkse 0, 0'dan büyükse değerin kendisi geri döndürülür.

> Sigmoid: Applies the sigmoid activation function. For small values (<-5), sigmoid returns a value close to zero, and for large values (>5) the result of the function gets close to 1. (Keras'ın web sitesinden)


## Output Layer Aktivasyonları:

Bu tarz aktivasyon fonksiyonları output layerlarda kullanılır.

> Linear: Adı üstünde, hangi value girerse o çıkar. Aynı zamanda aktivasyon fonksiyonu yok da denebilir. Eğer bir fonksiyona giren sayı olduğu gibi çıkıyorsa anlamı yoktur.

> Sigmoid: Hidden layer activation fonksiyonu olan sigmoid'le aynıdır.

> Softmax: En çok kullandığımız önemli bir aktivasyon fonksiyonudur. Softmax fonksiyonunu en güzel şu örnekle açıklayabilirim: Fashion mnist dataseti'nde 10 tane olağan output var. Softmax fonksiyonu, bir array içinde, 10 tane sayı verir. Bu sayıların toplamı her zaman 1'e eşit olur. En yüksek rakam, en olanak outputtur.

## Çok fazla aktivasyon var. Hangisini hangi durumda seçmem lazım?

https://machinelearningmastery.com/wp-content/uploads/2020/12/How-to-Choose-an-Output-Layer-Activation-Function.png

````python
def relu(num):
  return max(num, 0.0)

def sigmoid(num):
  return 1.0 / (1.0 + np.exp(-num))

def softmax(num):
	return np.exp(num) / np.exp(num).sum()

values = np.arange(-10, 10, 0.01)
````

````python
plt.plot(values, [relu(num) for num in values])
plt.title("Relu Fonksiyonu")
````

````python
plt.plot(values, [sigmoid(num) for num in values])
plt.title("Sigmoid Fonksiyonu")
````

````python
softmax_array = np.array([1, 2, 3])
print(f"Output: {softmax(softmax_array)}")
print(f"Toplam: {softmax(softmax_array).sum()}")
````

## Binary Classification:

> A greater value of entropy for a probability distribution indicates a greater uncertainty in the distribution. Likewise, a smaller value indicates a more certain distribution. This makes binary cross-entropy suitable as a loss function – you want to minimize its value. We use binary cross-entropy loss for classification models which output a probability p.

## Multiclass Classification:

Cross Entropy:

> Cross-entropy will calculate a score that summarizes the average difference between the actual and predicted probability distributions for all classes in the problem. The score is minimized and a perfect cross-entropy value is 0.
