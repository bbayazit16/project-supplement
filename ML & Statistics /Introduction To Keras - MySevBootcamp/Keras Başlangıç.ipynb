# Neden Keras?

Keras is an API designed for human beings, not machines. Keras follows best practices for reducing cognitive load: it offers consistent & simple APIs, it minimizes the number of user actions required for common use cases, and it provides clear and actionable feedback upon user error.

This makes Keras easy to learn and easy to use. As a Keras user, you are more productive, allowing you to try more ideas than your competition, faster -- which in turn helps you win machine learning competitions.

This ease of use does not come at the cost of reduced flexibility: because Keras integrates deeply with low-level TensorFlow functionality, it enables you to develop highly hackable workflows where any piece of functionality can be customized.

Keras has broad adoption in the industry and the research community
With over 400,000 individual users as of early 2021, Keras has strong adoption across both the industry and the research community. Together with TensorFlow 2, Keras has more adoption than any other deep learning solution -- in every vertical.

You are already constantly interacting with features built with Keras -- it is in use at Netflix, Uber, Yelp, Instacart, Zocdoc, Square, and many others. It is especially popular among startups that place deep learning at the core of their products.

Keras & TensorFlow 2 are also a favorite among researchers, coming in #1 in terms of mentions in scientific papers indexed by Google Scholar. Keras has also been adopted by researchers at large scientific organizations, such as CERN and NASA.

## Preprocessing

Preprocessing, yapay zeka'ya veriyi vermeden önce yapmanız gerekenlerdir. 

Python ve Yapay zekalar küçük rakamlarda daha iyi çalışır. Bu sebeple besleyeceğimiz verileri 0 ile 1 arasında tutmak iyi bir yöntemdir.

Biz de her resmi 255'e bölebiliriz. Grayscale  değerleri 0 ile 255 arasındadır. Yapay zekaya besleyeceğimiz resimleri 0 ile 1 arasına almamız için 255'e bölebiliriz.

Bunun yanında keras.utils.normalize(ARRAY) gibi bir fonksiyon var, ancak buna şimdilik girmeyeceğiz.

Her zaman değerler 0 ile 1 arasına getirilmek için 255'e bölünemez. Bunun için ayrıca MinMaxScaler adlı bir fonksiyon vardır.

`````python
x_train = x_train / 255
x_test = x_test / 255
# numpy arraylerine işlem uygulanabilir.
`````

## Keras Modelleri

Keras, kendilerinin de söyledikleri gibi, makineler için değil, insanlar için tasarlanmış bir api. Yani anlaması kolay olmalı.

Keras ile Sequential bir model kolayca açılabilir. Sequential model, keras'ın sitesine göre:

```
A Sequential model is appropriate for a plain stack of 
layers where each layer has exactly one input tensor and one output tensor.
```

ya da, kısacası,

```
A model which has linear stacks of layers.
```

## Sequential model nedir anlamadım?

Aşağıdaki resmi görünce anlayıcağınza eminim:

resized dense layer sequential model.jpg

Sequential modeller, bir list içinde layer alırlar. Bunlardan en sık kullanılan, dense layerlerdır. Bu:

````
The dense layer is a neural network layer that is connected 
deeply, which means each neuron in the dense layer receives
input from all neurons.
````

Dense layerlar dışında, MaxPool2D, Conv2D ve Dropout layerlar bulunur.

````
The dense layer is a neural network layer that is connected
deeply, which means each neuron in the dense layer receives
input from all neurons.

Dropout: Apply dropout to the model, setting a fraction
of inputs to zero in an effort to reduce over fitting.

MaxPool2D: Pooling layerları. İlerde geri dönüş yapacağım.
Conv2D: Convolution layerları. Buna da ilerde geri döneceğim.
````

##  Önemli bir uyarı
Keras layerlarında en sondaki layer output sayınıza eşit olmalı. Örneğin Fashion MNIST datasetinde 10 tane olanak (possible output) var. O zaman son layerımız 10 nöronlu bir dense layer olmalı.

`````python
plt.imshow(((x_train[0].reshape(784)).reshape((28, 28))) * 255)
`````

`````python
model = keras.Sequential([
                          keras.layers.Flatten(),
                          keras.layers.Dense(32),
                          keras.layers.Dense(10, activation='softmax')
])
`````

1.Layer:

> keras.layers.Flatten() ile resmimizi yapay zekaya besledik. Flatten layerı resmi düzleştirir.

image.png

Not: input_shape ve aktivasyon zorunlu parametreler değildir. Eğer aktivasyonu boş bırakırsanız linear aktivasyon olarak otomatik ayarlanır. Input shape de modeli asıl eğitmeye başladığınız zaman otomatik olarak belirlenir. Input_shape'i baştan belirlemenin bir iyi yanı modeli özetleyebilmemiz veya hata olup olmadığını anlayabilmemizdir.

---

2.Layer:

> Input layer'dan sonra, 32 nöronlu bir dense layer açtık.  Bu layerda RELU aktivasyonu bulunmakta. Bu model'in alacağı input ise x_train olacağı için x_train'in numpy shape'ini verdik. (60000, 28, 28) 

---

3.Layer:
> Softmax aktivasyonlu, 10 tane nöronlu bir Dense layer belirledik.








`````python
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# model'imizin optimizer'ını Adam olarak seçelim. Loss ise sparse categorical crossentropy olsun.
`````

`````python
model.fit(x_train, y_train, epochs=5)
# model'i asıl eğittiğimiz satır.
# x_train input, y_train output. 
# epoch: 10 kere bu verilerin üstünden geç.
`````

`````python
model.evaluate(x_test, y_test)
# model yüzde kaç oranında çalışıyor? x_test'i y_test'le karşılaştır ve yüzdeyi ver.
`````

`````python
predictions = model.predict(x_test)
`````

`````python
plt.imshow(x_test[1])
np.argmax(predictions[1])
# model 9 olarak buluyor. 9 da ankle boot'a eşit.
`````

`````python
model.save("")
`````
